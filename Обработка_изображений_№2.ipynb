```python
import numpy as np
import imutils
import cv2
from google.colab.patches import cv2_imshow
```

```python
image_1 = cv2.imread('/content/1.jpg')
image_2 = cv2.imread('/content/2.jpg')
image_3 = cv2.imread('/content/3.jpg')
image_4 = cv2.imread('/content/4.jpg')
image_5 = cv2.imread('/content/5.jpg')
processed_image = cv2.medianBlur(image_1, 3)
cv2.imwrite('processed_image_1_m.png', processed_image)
processed_image = cv2.medianBlur(image_2, 3)
cv2.imwrite('processed_image_2_m.png', processed_image)
processed_image = cv2.medianBlur(image_3, 3)
cv2.imwrite('processed_image_3_m.png', processed_image)
processed_image = cv2.medianBlur(image_4, 3)
cv2.imwrite('processed_image_4_m.png', processed_image)
processed_image = cv2.medianBlur(image_5, 3)
cv2.imwrite('processed_image_5_m.png', processed_image)
```

```python
image_1 = cv2.imread('/content/1.jpg')
image_2 = cv2.imread('/content/2.jpg')
image_3 = cv2.imread('/content/3.jpg')
image_4 = cv2.imread('/content/4.jpg')
image_5 = cv2.imread('/content/5.jpg')

kernel = np.ones((3,3),np.float32)/9

processed_image = cv2.filter2D(image_1,-1,kernel)
cv2.imwrite('processed_image_1.png', processed_image)
processed_image = cv2.filter2D(image_2,-1,kernel)
cv2.imwrite('processed_image_2.png', processed_image)
processed_image = cv2.filter2D(image_3,-1,kernel)
cv2.imwrite('processed_image_3.png', processed_image)
processed_image = cv2.filter2D(image_4,-1,kernel)
cv2.imwrite('processed_image_4.png', processed_image)
processed_image = cv2.filter2D(image_5,-1,kernel)
cv2.imwrite('processed_image_5.png', processed_image)
```

```python
image = cv2.imread("/content/custom.jpg") # Путь к исходному изображению
fragment = cv2.imread("/content/Et_custom.jpg") # Путь к фрагменту изображения (эталону)

(fragmentHeight, fragmentWidth) = fragment.shape[:2] # Осуществляем поиск фрагмента на изображении
print(fragmentHeight, fragmentWidth)
result = cv2.matchTemplate(image, fragment, cv2.TM_CCOEFF)

(_, a, minLoc, maxLoc) = cv2.minMaxLoc(result) # Выделяем искомый фрагмент и отделяем его от изображения

print(_, a, minLoc, maxLoc)
topLeft = maxLoc

botRight = (topLeft[0] + fragmentWidth, topLeft[1] + fragmentHeight)

roi = image[topLeft[1]:botRight[1], topLeft[0]:botRight[0]]

mask = np.zeros(image.shape, dtype="uint8")
image = cv2.addWeighted(image, 0.30, mask, 0.75, 0)

image[topLeft[1]:botRight[1], topLeft[0]:botRight[0]] = roi

cv2_imshow(imutils.resize(image, height=950))
```
